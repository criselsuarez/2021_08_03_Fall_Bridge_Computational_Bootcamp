```python
years = data.columns.str.strip('gdpPercap_')
data.columns = years.astype(int)
```

```python
data
```

```python
data.loc['Australia'].plot()
```

```python
%matplotlib inline
```

```python
data.T.plot()
plt.ylabel('GDP')
```

```python
plt.style.use('ggplot')
data.T.plot(kind='bar')
plt.ylabel('GDP')
```

```python
year= data.columns
gpd_aus = data.loc['Australia']
gpd_nz = data.loc['New Zealand']


plt.plot(year, gpd_aus, 'g:', label = 'Australia')
plt.plot(year, gpd_nz, 'r:', label = 'New Zealand')

plt.legend(loc = 'lower right')
plt.savefig('aus_nz_gdp.jpg')
plt.ylabel('GDP')
plt.xlabel('Years')
plt.title('Oceania')
plt.yscale('log')
```
